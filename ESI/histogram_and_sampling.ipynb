```python
import os
import sys
if os.path.realpath('..') not in sys.path: 
  sys.path.append(os.path.realpath('..'))
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_teor = pd.read_pickle(os.path.join('..','data','calculated_data.pkl'))
```

```python
x_axis_label = "Molecular Assembly"
y_axis_label = "Count"

fig, ax = plt.subplots(figsize=(6,4))
sns.histplot(df_teor.assembly_index,binwidth=1)
ax.set_xlabel(x_axis_label, size = 10 )
ax.set_ylabel(y_axis_label, size = 10)
ax.set_xticks(list(range(0,23,2)))
fig.show()
```
